```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt

# Meshgrid
x, y = np.meshgrid(np.linspace(0, 1, 25),np.linspace(0, 1, 25))

# Directional vectors
u =- x*(1-x+3*y)/(((x*(1-x+3*y))**2+((3+x-3*y)*y)**2)**0.5)
v = (3+x-3*y)*y/(((x*(1-x+3*y))**2+((3+x-3*y)*y)**2)**0.5)

# Plotting Vector Field with QUIVER
plt.quiver(x, y, u, v, color='r')
plt.title('Phase Sapace')

# Setting x, y boundary limits
plt.xlim(-0.04, 1)
plt.ylim(-0.04,1)

# phase orbits for different initial condition

# 1st initial condition

x0=[0.95]
y0=[0.00000001]
dt=0.01

for i in range(0,1000,1):
    xn=-x0[i]*(1-x0[i]+3*y0[i])*dt+x0[i]
    yn=(3+x0[i]-3*y0[i])*y0[i]*dt + y0[i]
    x0.append(xn)
    y0.append(yn)
plt.plot(x0,y0)

#2nd initial condition
x1=[0.9]
y1=[0.00001]
dt=0.01

for i in range(0,1000,1):
    xn=-x1[i]*(1-x1[i]+3*y1[i])*dt+x1[i]
    yn=(3+x1[i]-3*y1[i])*y1[i]*dt + y1[i]
    x1.append(xn)
    y1.append(yn)
plt.plot(x1,y1)


#3rd initial condition
x2=[0.999]
y2=[0.00000001]
dt=0.01

for i in range(0,1000,1):
    xn=-x2[i]*(1-x2[i]+3*y2[i])*dt+x2[i]
    yn=(3+x2[i]-3*y2[i])*y2[i]*dt + y2[i]
    x2.append(xn)
    y2.append(yn)
plt.plot(x2,y2)
#4th initial condition
x3=[0.9]
y3=[0.001]
dt=0.01

for i in range(0,1000,1):
    xn=-x3[i]*(1-x3[i]+3*y3[i])*dt+x3[i]
    yn=(3+x3[i]-3*y3[i])*y3[i]*dt + y3[i]
    x3.append(xn)
    y3.append(yn)
plt.plot(x3,y3)
plt.show()
```

# Cosh type potential

```python
import phaseportrait as pp
#autonomous ODE-s
def ode(x,y,z):
    return np.array([(-3/2)*x*(1 - x**2 + y**2) - np.sqrt(3/2)*y*y*z,(3/2)*y*(1 + x**2 - y**2) + np.sqrt(3/2)*x*y*z,np.sqrt(6)*x*(100 - z**2)])
phase_space = pp.PhasePortrait3D(ode, [(-1,1),(0,1),(-11,11)], xlabel = 'x', ylabel = 'y',zlabel = "z")
phase_space.plot()
```
