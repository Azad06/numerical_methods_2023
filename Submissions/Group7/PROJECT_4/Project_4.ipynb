```julia
# Finding critical points
# (0,0) , (0,1), (1,0)
```

```julia
function odesolve(x,y)
N  = 10000;
dt = 0.001;

xdata   = zeros(N); 
ydata   = zeros(N); 

xdata[1] = x
ydata[1] = y

for j=1:N-1
    
    xnew=x + dt*( -x*(1-x+3*y));
    ynew=y + dt*((3 + x âˆ’ 3*y)*y);
    
    xdata[j] = xnew
    ydata[j] = ynew
    
    x = xnew
    y = ynew
end

return xdata,ydata
end
```

```julia
function odesolve_new(x,y,z)
N  = 10000;
dt = 0.001;

xdata   = zeros(N); 
ydata   = zeros(N); 
zdata   = zeros(N);

xdata[1] = x
ydata[1] = y
zdata[1] = z
mu = 10
for j=1:N-1
    
    xnew=x + dt*( -1.5*x*(1-x^2+y^2)-sqrt(1.5)*y^2*z);
    ynew=y + dt*((1.5*y*(1+x^2-y^2)+sqrt(3/2)*x*y*z));
    znew=z + dt*(sqrt(6)*x*(mu^2-z^2))
    
    xdata[j] = xnew
    ydata[j] = ynew
    zdata[j] = znew
    
    x = xnew
    y = ynew
    z = znew
end

return xdata,ydata,zdata
end
```

```julia
xdata,ydata,zdata = odesolve_new(1,0.01,1.2)
plot(xdata,ydata,zdata)

xdata,ydata,zdata = odesolve_new(1,0.01,1.3)
plot!(xdata,ydata,zdata)
```
