## PROBLEM 7


Write a Julia Snippet which includes a user defined function, call it "initial_data.jl" to define a function on a domain
                                             x Ïµ [-10,10]
and generates a plot of the same.

```julia
# Importing necessary libraries
using Plots
using LaTeXStrings
```

```julia
# Defining initial function 
function initial_data(x)
    return exp(-x^2)
end
```

```julia
l = 10    # Defining range of x
dx = 0.01 # Defining step size

x = collect(-l:dx:l) # x data
y = initial_data.(x);
```

```julia
# Integrating Using Simpson Method
function Simpson(func,A, B ,N)
    h   = (B â A) / (N â 1) # step size
    sum = ( func(A)+ func(B) ) 
    
    for i = 1:2:N-1
        sum += 4*func(A+i *h)
    end
    
    for i = 2:2:N-2
        sum += 2*func(A+i *h)
    end
return h*sum/3
end
```

# PROBLEM 9

Write a Julia Snippet to generate a table giving co-efficients of the second derivative expansion in terms of the discretized function elements.

```julia
# Defining the function to calculate the coefficients in derivative expansion
function Coefficients(r)
M = zeros(r,r)
for i =1:r
    for j =1:r
        M[i,j] = 2 * (i)^(2*j) / factorial(2*j)
    end
end  
Mi=M^-1;
colm=Mi*ones(r)*(-2);
ck =Mi[1,:];
return cat(colm[1],ck,dims=1) #k=0 first
end
```

```julia
# Defining the r till where the table will be generated
r = 7

# Printing the table
for i=1:r
    print("r="*string(i)*"  ")
    println(["$v " for v in Coefficients(i)]...)
    #println("r="*string(i)*string(Coefficients(i)))
end
```
