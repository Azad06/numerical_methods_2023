```julia
#define function
function initial_data(x)
    return x^2-(3*sin(x))+(exp(x))-(x^4)
end

```

```julia
l=10
dx=.01
x = collect(-l:dx:l) 
y = initial_data.(x);
plot(x, y, title="Initial Function")
```

```julia
#Integrating using Trapezoid rule
function trapezoid(f,a,b,N) 
    h = (b-a)/N
    int = h * ( f(a) + f(b) ) / 2
    for k=1:N-1
        xk = (b-a) * k/N + a
        int = int + h*f(xk)
    end
    return int
end
trapezoid(initial_data,-l,l,10000)
```

```julia
function Coefficients(r)
M = zeros(r,r)
for i=1:r
    for j=1:r
        M[i,j] = 2*(i)^(2*j) / factorial(2*j)
    end
end
MI=M^-1;
column=MI*ones(r)*(-2);
ck=MI[1,:];
return cat(column[1],ck,dims=1)
end
```

```julia
for i=1:7
    print("r="*string(i)*"  ")
    println(["$v " for v in Coefficients(i)]...)
end
```
