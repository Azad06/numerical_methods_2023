```python
from phaseportrait import PhasePortrait3D, PhasePortrait2D
import math as mt
```

$\Lambda CDM $ Model without the field

${x'} = -x(1 - x + 3 y)$ <br />
${y'} = (3 + x - 3 y ) y $ <br />


```python
def lambda_cdm(x,y):
    return -x * (1 - x + 3 * y) , (3 + x - 3 * y ) * y
```

```python
Result1 = PhasePortrait2D(lambda_cdm , [[0 , 1], [0 , 1]],  xlabel='x(t)', ylabel='y(t)',
                             Title=r'$\Lambda$-CDM Phase Portrait' )
Result1.plot()
```

#### For cosh potential function (example given in the paper)
${V} = V_0 cosh(\mu X)$



${x'} = -{3 \over 2}x (1 - x^2 + y^2) - \sqrt{3\over 2} y^2 z$ <br />
${y'} = {3 \over 2}y (1 + x^2 - y^2) + \sqrt{3\over 2} x y z$ <br />
${z'} = \sqrt{6} x (\mu^2 - z^2) $ <br />


```python
mu_sq = 100

def lambda_cdm_cosh(y, x, z):
    return    1.5 * y * (1 + x*x - y * y) + mt.sqrt(3/2)*x*y*z, -1.5*x*(1- x*x + y*y ) - mt.sqrt(3/2) * y * y * z, mt.sqrt(6) *x * (mu_sq - z*z)
```

```python
Result = PhasePortrait3D(lambda_cdm_cosh , [ [0 , 1], [-1 , 1],[-10,10]], MeshDim=8, xlabel = 'y(t)', ylabel='x(t)', zlabel='z(t)',
                         Title=r'     $\Lambda$-CDM Scalar Field (cosh) Phase Portrait         ')
Result.plot()
```

#### Let's take a different potential 

#### For exp potential function
${V} = V_0 e^{+\mu X}$ <br />
#### or
$V_0 e^{-\mu X}$ <br />
#### which results in 
$f(z) = 0$ <br />
#### and therefore the ODE's are

${x'} = -{3 \over 2}x (1 - x^2 + y^2) - \sqrt{3\over 2} y^2 z$ <br />
${y'} = {3 \over 2}y (1 + x^2 - y^2) + \sqrt{3\over 2} x y z$ <br />
${z'} = 0$ <br />

```python
def lambda_cdm_exp(y, x, z):
    return    1.5 * y * (1 + x*x - y * y) + mt.sqrt(3/2)*x*y*z, -1.5*x*(1- x*x + y*y ) - mt.sqrt(3/2) * y * y * z, 0
```

```python
Result = PhasePortrait3D(lambda_cdm_exp , [ [0 , 1], [-1 , 1],[-10,10]], MeshDim=8, xlabel = 'y(t)', ylabel='x(t)', zlabel='z(t)',
                         Title=r'     $\Lambda$-CDM Scalar Field (exp) Phase Portrait         ')
Result.plot()
```

#### For cosh with some power potential function
${V} = V_0 [cosh(\mu X)]^p$<br />
#### which results in 
$f(z) = (p^2 \mu^2 - z^2)/ p $ <br />
$z = V_{,X} / V$ <br />
#### and therefore
$z = \mu p [tanh(\mu X)]$
#### z is bounded between
$(\mu p ,  -\mu p)$
#### and therefore the ODE's are

${x'} = -{3 \over 2}x (1 - x^2 + y^2) - \sqrt{3\over 2} y^2 z$ <br />
${y'} = {3 \over 2}y (1 + x^2 - y^2) + \sqrt{3\over 2} x y z$ <br />
${z'} = \sqrt{6} x (p^2 \mu^2 - z^2)/ p $

```python
# taking the value of p = 4


def lambda_cdm_cosh_pow(y, x, z):
    return    1.5 * y * (1 + x*x - y * y) + mt.sqrt(3/2)*x*y*z, -1.5*x*(1- x*x + y*y ) - mt.sqrt(3/2) * y * y * z, mt.sqrt(6)*x*(16*mu_sq - z*z)/4
```

```python
Result = PhasePortrait3D(lambda_cdm_cosh_pow , [ [0 , 1], [-1 , 1],[-20,20]], MeshDim=8, xlabel = 'y(t)', ylabel='x(t)', zlabel='z(t)',
                         Title=r'     $\Lambda$-CDM Scalar Field (cosh to power 4)  Phase Portrait         ')
Result.plot()
```
