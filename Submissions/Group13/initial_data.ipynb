```julia
#Exercise-1 (Defining & Plotting a function over the domain [-10,10]

using Plots
f(x) = (x^2)*exp(-x^2) 
i = -10.0:0.1:10.0
fi = []
for j in i
    push!(fi,f.(j))
end
plot(i, fi, ylabel = "y", xlabel = "x", label = "(x^2)*exp(-x^2)")
```
