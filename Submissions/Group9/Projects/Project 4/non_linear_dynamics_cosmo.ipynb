```python
from matplotlib import pyplot as plt
import numpy as np
import phaseportrait as pp
```

### $\Lambda$-CDM Model

```python
# while defining ODEs we use the definition of derivatives and return them
def lcdm_odes(x,y):
    # x for radiation and y for dark matter
    # 1 - x - y for normal matter
    # x, y = v
    return np.array([-x*(1 - x +3*y), y*(3 + x - 3*y)])


lcdm_pp = pp.PhasePortrait2D(lcdm_odes, [(0,1), (0,1)], xlabel='x(t)', ylabel='y(t)',
                             Title=r'$\Lambda$-CDM Phase Portrait')
lcdm_pp.plot()
```

### $\Lambda$-CDM Scalar Field Model

#### 1. cosh potential



```python
%matplotlib widget
# while defining ODEs we use the definition of derivatives and return them
mu = 10
def lcdm_scalar_field_cosh_potential_odes(x,y,z):
    return np.array([(-3/2)*y*(1 - x**2 + y**2) - np.sqrt(3/2)*y*y*z,
                     (3/2)*y*(1 + x**2 - y**2) + np.sqrt(3/2)*x*y*z,
                     np.sqrt(6)*x*(mu**2 - z**2)])

lcdm_sf_cosh_pp = pp.PhasePortrait3D(lcdm_scalar_field_cosh_potential_odes, [(-1,1),(0,1),(-1.1*mu,1.1*mu)], xlabel = 'x(t)', ylabel = 'y(t)', 
                             zlabel = 'z(t)', MeshDim = 8
                             ,Title=r'$\Lambda$-CDM Scalar Field (cosh) Phase Portrait')
lcdm_sf_cosh_pp.plot()
```
